# Question 1

Change the original task into a classification task. 

* Create a new feature called price_class. 

* For prices higher than $113,000, price_class equals 1.

* For prices that lower or equal to $113,000 price_class is 0. 

* Print the first five rows of the table (already in precode).

**Hint:** Use the loc() function to access elements by specifying their row and column.


```python
import pandas as pd

df = pd.read_csv('https://bit.ly/TrainDataUS')

# < write code here >

df.head()
```

### Solution

```python
import pandas as pd

df = pd.read_csv('https://bit.ly/TrainDataUS')

df.loc[df['last_price'] > 113000, 'price_class'] = 1
df.loc[df['last_price'] <= 113000, 'price_class'] = 0

print(df.head())
```

### Solution

```python
import pandas as pd

df = pd.read_csv('https://bit.ly/TrainDataUS')

df.loc[df['last_price'] > 113000, 'price_class'] = 1
df.loc[df['last_price'] <= 113000, 'price_class'] = 0

features = df.drop(['last_price', 'price_class'], axis=1)
target = df['price_class']

print(features.shape)
print(target.shape)
```
